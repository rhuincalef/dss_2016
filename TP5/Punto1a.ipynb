```python
from sklearn import preprocessing
from sklearn import datasets
import numpy as np
```

```python
iris_mean_removal = datasets.load_iris().data
iris_mean_removal
```

```python
iris_mean_removal_procesado = preprocessing.scale(iris_mean_removal)
```

#### Maximos y minimos despues de preprocesar los datos

```python
print "Valor minimo: %s", iris_mean_removal_procesado.min()
print "Valor maximo: %s",iris_mean_removal_procesado.max()
```

#### Desvio estandar y media despues de filtrar los datos 

```python
#Media y desvio estandar de la media removida
print "Media cercana a cero %s", iris_mean_removal_procesado.mean(axis=0)
print "Desvio estandar de 1 %s",iris_mean_removal_procesado.std(axis=0)
```
